# Function of calculating Confusion Matrix with Accuracy, Precision, Recall & F1-Score Calculation

```python
def fnConfusionMatrix(y_test, y_pred , algo):
    
    cm = confusion_matrix(y_test, y_pred)
    print ("ALGORITHM : " , algo)
    print ("--------------------")
    print ("Confusion Matrix : \n", cm)
    print ("Accuracy : ", accuracy_score(y_test, y_pred)*100 , "%")
    print ("Recall : ", recall_score(y_test, y_pred,average="macro")*100 , "%")
    print ("Precision : ", precision_score(y_test, y_pred,average="macro")*100 , "%")
    print ("F1-Score : ", f1_score(y_test, y_pred,average="macro")*100 , "%") 
```

```python
def fnPredict(classifier, X_test, y_test):
    y_pred = classifier.predict(X_test)
    return y_pred
```
